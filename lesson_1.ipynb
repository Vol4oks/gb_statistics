### Пример 1

Моделируем 60-ти кратное подбрасование игральной кости. 
Событием А будем считать выыподение числа 3 
и найдем его относительною частоту

```python
import numpy as np
np.random.seed(1)
n = 60
b = np.random.randint(1, 7, n)
b
```

Вычислим мощность подможества, где в результате испытания  
выпадало число 3, то есть наблюдалось событие А

```python
a=b[b==3]
a
```

```python
m = len(a)
m
```

Вычисляем относительною частоту события А

```python
M = m / n
M
```

## Комбинаторика

### Пример 1
Сколькими способами можно выбрать из колоды, состоящий из 36 карт, 4 карты?

```python
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
combinations(36, 4)
```

### Пример 2
В магазине 20 покупателей. Сколькими способами они могут образовать очередь из 5 человек?

```python
def arrangements(n, k):
    return int(factorial(n) / factorial(n -k))
```

```python
arrangements(20, 5)
```

### Пример 3

### Пример 4
Из колоды, состоящей из 36 карт случайным образом выбраны 5. Сколькими способами можно выбрать эти карты так, чтобы среди них оказалось от 2 до 3 Тузов 

```python
# Количество комбинаций только Тузов. С 2 -тузами и с 3 -тузами отдельно
C_1_t = combinations(4, 2)
C_2_t = combinations(4, 3)
C_1_t, C_2_t
```

```python
# Количество комбинаций без Тузов. С 2 -тузами и с 3 -тузами отдельно
C_1_n = combinations(32, 3)
C_2_n = combinations(32, 2)
C_1_n, C_2_n
```

```python
# Количество комбинаций . С 2 -тузами и с 3 -тузами отдельно
C_1 = C_1_t * C_1_n
C_2 = C_2_t * C_2_n
C_1, C_2
```

```python
# Всего комбинаций 
C = C_1 + C_2
print(f' ОТВЕТ: Всего возможно {C} - комбинаций')
```

### Пример 5
Из колоды 36 карт случайным образом берут 5 карт.
Найти вероятность того, среди 5 карт будет 2 туза

```python
# Находим количество нужных исходов
C_1_t = combinations(4, 2)
C_1_n = combinations(32, 3)
C_1 = C_1_t * C_1_n
C_1
```

```python
# Находим общее число исходов 
C = combinations(36, 5)
C
```

```python
# Находим вероятность того, среди 5 карт будет 2 туза
P = C_1 / C
print(f'ОТВЕТ: Вероятность того, что среди 5 карт, взятых из колоды в 36, \
будет 2 туза - {round(P, 4)} или около {round(P*100)} %')
```
