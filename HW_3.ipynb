#### Задание 1
Даны значения зарплат из выборки выпускников: 

100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. 

Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
v = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
```

```python
# Среднее арифметическое
sum_v = 0
for i in v:
    sum_v += i
mean_v = sum_v / len(v)
print(f'Среднее арифметическое: {mean_v}')
```

```python
# Смещенная дисперсия
ds_v = 0
for i in v:
    ds_v += (i - mean_v)**2
ds_s = ds_v/len(v)
print(f'Смещенная дисперсия: {ds_s}')
```

```python
# Несмещенная дисперсия
ds_v = 0
for i in v:
    ds_v += (i - mean_v)**2
ds_n = ds_v/(len(v)-1)
print(f'Несмещенная дисперсия: {round(ds_n, 2)}')
```

```python
# Среднее квадратичное отклонение
ot_v = ds_n**0.5
print(f'Среднее квадратичное отклонение: {round(ot_v, 2)}')
```

#### Задание 2
В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
from math import factorial
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# из 1 ящика 2 белых мяча и из 2 ящика 1 белый мяч
p1 = combinations(5, 2) / combinations(8, 2) * combinations(5, 1) * combinations(7, 3) / combinations(12, 4)
# из 1 ящика 1 белый мяч и из 2 ящика 2 белых мяча
p2 = combinations(5, 1) * combinations(3, 1) / combinations(8, 2) * combinations(5, 2) * combinations(7, 2) / combinations(12, 4)
# из 1 ящика 0 белых мяча и из 2 ящика 3 белых мяча
p3 = combinations(3, 2) / combinations(8, 2) * combinations(5, 3) * combinations(7, 1) / combinations(12, 4)
p = p1 + p2 + p3
print(f'Вероятность, что 3 мяча белые = {round(p*100, 2)}%')
```

#### Задание 3
На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
p = 1/3 * 0.9 + 1/3 * 0.8 + 1/3 * 0.6
p1 = 1/3 * 0.9 / p
p2 = 1/3 * 0.8 / p
p3 = 1/3 * 0.6 / p
print(f'Вероятность, что выстрел произведен первым спортсменом = {round(p1*100, 2)}%')
print(f'Вероятность, что выстрел произведен вторым спортсменом = {round(p2*100, 2)}%')
print(f'Вероятность, что выстрел произведен третьим спортсменом = {round(p3*100, 2)}%')
```

#### Задание 4
В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
p = 1/4 * 0.8 + 1/4 * 0.7 + 2/4 * 0.9
pA = 1/4 * 0.8 / p
pB = 1/4 * 0.7 / p
pC = 2/4 * 0.9 / p
print(f'Вероятность, что первую сессию сдал студент факультета A = {round(pA*100,2)}%')
print(f'Вероятность, что первую сессию сдал студент факультета B = {round(pB*100,2)}%')
print(f'Вероятность, что первую сессию сдал студент факультета C = {round(pC*100,2)}%')
```

#### Задание 5
Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
p1 = 0.1
p2 = 0.2
p3 = 0.25
q1 = 1 - p1
q2 = 1 - p2
q3 = 1 - p3
P_1 = p1 * p2 * p3
P_2 = p1 * p2 * q3  +  p1 * q2 * p3  +  q1 * p2 * p3
P_3 = 1 - q1 * q2 * q3
P_4 = p1 * q2 * q3 + q1 * p2 * q3  +  q1 * q2 * p3 + P_2
print(f'Вероятность, что в первый месяц выйдут из строя все детали = {round(P_1*100,2)}%')
print(f'Вероятность, что в первый месяц выйдут из строя только две детали = {round(P_2*100,2)}%')
print(f'Вероятность, что в первый месяц выйдут из строя хотя бы одна деталь = {round(P_3*100,2)}%')
print(f'Вероятность, что в первый месяц выйдут из строя от одной до двух деталей = {round(P_4*100,2)}%')
```
