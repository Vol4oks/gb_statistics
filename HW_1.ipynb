#### Задание 1
Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

```python
from math import factorial
# функция для вычисления комбинации из урока
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# A)
# расчитаем количество благоприятных исходов  и всего исходов
C_KR = combinations(13, 4)
C_all = combinations(52, 4)
C_KR, C_all
```

```python
# Расчитываем вероятность
C = C_KR / C_all
print(f'Вероятность, что все карты - крести: {round(C, 5)}')
```

```python
# Б)
# Расчитаем вероятность того, что в комбинации не окажется туза
C_n = combinations((52-4), 4)
C_ = C_n / C_all
C_
```

```python
# Расчитаем вероятность хотя бы одного туза
C = 1 - C_
print(f'Вероятность, что среди 4-х карт окажется хотя бы один туз: {round(C, 5)} или около {round(C*100)} %')
```

#### Задание 2
На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

```python
# функция для вычисления комбинации из урока
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# Находим количество всех событий
C = combinations(10, 3)
C
```

```python
# Находим вероятность выбора нужного сочитания 
P = 1 / C
print(f'вероятность того, что человек, не знающий код,\
откроет дверь с первой попытки - {round(P, 4)} или {round(P*100, 2)}%')
```

#### Задание 3
В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

```python
# функция для вычисления комбинации из урока
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# Находим количество всех событий
C = combinations(15, 3)
C
```

```python
# расчитаем количество благоприятных исходов 
C_1 = combinations(9, 3)
C_1
```

```python
# Находим вероятность выбора нужного сочитания 
P = C_1/ C
print(f'вероятность того, что все извлеченные детали окрашены \
- {round(P, 4)} или {round(P*100, 1)}%')
```

#### Задание 4
В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

```python
# функция для вычисления комбинации из урока
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# расчитаем количество благоприятных исходов 
C_1 = combinations(2, 2)
C_1
```

```python
# Находим количество всех событий
C = combinations(100, 2)
C
```

```python
# Находим вероятность выбора нужного сочитания 
P = C_1 / C
P
```

```python
print(f'вероятность того, что 2 приобретенных билета \
окажутся выигрышными - {round(P, 4)} или {round(P*100, 3)}%')
```
