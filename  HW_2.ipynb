#### Задание 1
Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
# используем готовую функцию формулы Бернулли
# где попаданий в цель - 85
# выстрелов - 100
# вероятность - 0.8

P = stats.binom.pmf(85, 100, 0.8)
print(f'Вероятность того, что стрелок попадет в цель\
ровно 85 раз равна: {round(P, 4)} или {round(P, 4) * 100} %')
```

#### Задание 2
Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? 

```python
# используем готовую функцию формулы Бернулли
# количество перегоревших ламп - 0
# общее количество ламп - 5000
# вероятность перегорания - 0.0004

P = stats.binom.pmf(0, 5000, 0.0004)
print(f'Вероятность, что ни одна из них не перегорит\
 в первый день: {round(P, 3)} или {round(P * 100, 1) } %')
```

```python
# используем готовую функцию формулы Бернулли
# количество перегоревших ламп - 2
# общее количество ламп - 5000
# вероятность перегорания - 0.0004

P = stats.binom.pmf(2, 5000, 0.0004)
print(f'Вероятность, что перегорят ровно две: {round(P, 3)} или {round(P * 100, 1) } %')
```

#### Задание 3

#### Задание 4
В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? 

```python
from math import factorial
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# находим количество положительных исходов в первом ящике
n = 7
k = 2
a = combinations(n, k)
print(f'Количество положительных исходов: {a}')
```

```python
# находим общее количество исходов в первом ящике
n = 10
k = 2
C_a = combinations(n, k)
print(f'Количество положительных исходов: {C_a}')
```

```python
# находим количество положительных исходов во втором ящике
n = 9
k = 2
b = combinations(n, k)
print(f'Количество положительных исходов: {b}')
```

```python
# находим общее количество исходов во втором ящике
n = 11
k = 2
C_b = combinations(n, k)
print(f'Количество положительных исходов: {C_b}')
```

```python
# находим вероятность того, что все мячи белые
Pa = (a/C_a) * (b/C_b)
print(f'Вероятность того, что все мячи белые равна: {round(Pa, 3)} или {round(Pa * 100, 1) } %' )
```
