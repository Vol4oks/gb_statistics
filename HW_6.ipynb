#### Задание 1
Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):

zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],

ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].

Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy
Полученные значения должны быть равны.
Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков,
а затем с использованием функций из библиотек numpy и pandas.

```python
import numpy as np
import pandas as pd
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
cov=np.mean(zp*ks)-np.mean(zp)*np.mean(ks)
print(f'ковариация элемент действиями: {round(cov, 2)}' )
print(f'ковариация: {round(np.cov(zp,ks,ddof=0)[1,0], 2)}')


print(np.std(zp))
print(np.std(ks))

df=pd.DataFrame({'zp': zp, 'ks': ks})
print(f'коэффициент корреляции np     ',np.corrcoef(zp,ks,ddof=0)[1,0])
print(f'коэффициент корреляции Пирсона',cov/(np.std(zp)*np.std(ks)))
print(f'коэффициент корреляции pd     ',df.corr()['ks'][0])
```
